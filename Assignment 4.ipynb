1. Difference between a spreadsheet and a database
Spreadsheet: A file (like Excel) where data is stored in rows and columns. Best for small datasets, calculations, and quick analysis.
Database: A structured system that stores large amounts of data in multiple related tables. Designed for efficiency, accuracy, and multi-user access.

Key difference:
Spreadsheets are simple and flexible; databases are powerful and structured for large-scale data management.

2. Advantages & Disadvantages of Spreadsheets (Health Data)
Advantages
Easy to use (no technical skills needed)
Good for small datasets (e.g., clinic patient logs)
Built-in formulas and charts
Quick data entry and visualization
Disadvantages
Prone to human error (manual entry mistakes)
Limited security (risk for sensitive health data)
Not ideal for large datasets
Difficult for multiple users at once
Poor data integrity (duplicates, inconsistencies)


3. Advantages & Disadvantages of Databases (Health Data)
Advantages
Handles large, complex datasets efficiently
Strong security (important for patient confidentiality)
Supports multiple users simultaneously
Maintains data integrity (less duplication, more accuracy)
Can link related data (patients, visits, treatments)
Disadvantages
Requires technical knowledge
More expensive to set up and maintain
Less flexible for quick, informal analysis
Needs structured design before use

4. Tasks using a Spreadsheet
Creating charts/graphs (e.g., infection rates)
Budget tracking for a clinic
Calculating averages (e.g., patient age)
Simple data entry lists (e.g., vaccination records)


5. Tasks using a Database
Managing electronic health records (EHRs)
Tracking patient histories across visits
Running complex queries (e.g., patients with diabetes + hypertension)
Managing hospital systems (appointments, billing)


6. Tasks using BOTH Spreadsheet and Database
Data analysis (basic vs advanced)
Reporting (e.g., monthly patient counts)
Data storage (small vs large scale)
Sorting and filtering data


7. Tasks using NEITHER
Performing surgery
Diagnosing patients directly
Conducting physical exams
Providing patient counseling

(These require human interaction, not data tools.)

8. Tasks using Spreadsheet but NOT Database
Quick one-time calculations
Creating simple charts instantly
Ad-hoc data exploration without setup
Personal tracking (e.g., daily calorie log)


9. Tasks using Database but NOT Spreadsheet
Managing millions of patient records
Enforcing strict data security rules (HIPAA-level systems)
Handling multiple users editing data at the same time
Complex relational queries across multiple tables


10. (Repeat) Tasks using BOTH Spreadsheet and Database
Data entry
Basic analysis
Sorting/filtering
Generating reports
